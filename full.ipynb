First, you need to import NumPy and OpenCV so that you can use them.

```python
import numpy as np
import cv2
```

Feel free to ignore the following code, it allows you to show OpenCV images in this environment

```python
# Notebook setup + convenience functions
%matplotlib inline
import matplotlib.pyplot as plt

def imshow(img):
    '''Function to show image, because matplotlib and OpenCV aren't a perfect match'''
    if len(img.shape) == 2 or img.shape[2] == 1:
        # grayscale image
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
    else:
        # 'Color' image
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

OpenCV code to load an image and display it in the notebook

```python
img = cv2.imread('images/2016-p2.jpg')
imshow(img)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
imshow(hsv)
```

```python
lower = np.array([0, 145, 80])
upper = np.array([255, 255, 255])

filtered = cv2.inRange(hsv, lower, upper)
imshow(filtered)
```
